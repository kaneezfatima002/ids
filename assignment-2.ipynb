```python
#task1
users = [
    (1, "Alice", 25, "USA"),
    (2, "Bob", 32, "Canada"),
    (3, "Charlie", 28, "USA"),
    (4, "David", 35, "UK"),
    (5, "Eve", 40, "Canada"),
    (6, "Frank", 29, "USA"),
    (7, "Grace", 22, "Canada"),
    (8, "Hannah", 33, "USA"),
    (9, "Isabella", 31, "Canada"),
    (10, "Jack", 38, "USA"),
    (11, "Eve", 34, "Canada")  # Duplicate name example
]

def filter_user(user_tuple):
    filtered=[]
    for user in user_tuple:
        if user[2] > 30 and (user[3]=="USA" or user[3]=="Canada" ):
            filtered.append(user[1])
    return filtered

print("Users older than 30 yeasr and living in USA or Canada:")
print(filter_user(users))
print()


def age(user):
    return user[2]

def top_10_users(usr_lst):
    sorted_list = sorted(usr_lst, key=age, reverse=True)
    top_users = sorted_list[:10]
    return top_users

print("Top 10 oldest users:")
print( top_10_users(users))


def dupicate_names(usr_lst):
    duplicates=[]
    unique= set()
    for usr in usr_lst:
        name= usr[1]
        if name in unique:
            duplicates.append(name)
        else:
            unique.add(name)
    return duplicates

print()
print("Duplicate User's Name")
print(dupicate_names(users))


```

```python
#task2
transactions = [
    (1, 101, 150.0, '2024-10-01 10:00:00'),
    (2, 102, 200.0, '2024-10-02 12:30:00'),
    (3, 101, 250.0, '2024-10-03 09:15:00'),
    (4, 103, 300.0, '2024-10-04 14:45:00'),
    (5, 102, 400.0, '2024-10-05 08:30:00'),
    (6, 104, 500.0, '2024-10-06 11:00:00'),
    (7, 105, 600.0, '2024-10-07 15:15:00'),
    (8, 101, 700.0, '2024-10-08 13:30:00'),
    (9, 106, 800.0, '2024-10-09 17:00:00'),
    (22, 116, 2100.0, '2024-10-22 15:30:00'),
    (23, 117, 2200.0, '2024-10-23 11:15:00'),
    (24, 118, 2300.0, '2024-10-24 08:00:00'),
    (25, 119, 2400.0, '2024-10-25 14:45:00'),
    (26, 120, 2500.0, '2024-10-26 10:30:00')
]

def total_unique_users(trans_lst):
    count=0
    unique=[]
    for transaction in trans_lst:
        usr= transaction[1]
        if usr not in unique:
            unique.append(usr)
            count+=1
    return count

print(f"Unique User Count:{total_unique_users(transactions)}")



def hghest_amount_trans(trans_lst):
    highest_id= trans_lst[0]
    for transaction in trans_lst:
        if transaction[2]> highest_id[2]:
            highest_id = transaction
    return highest_id

print()
print("User with Hghest Transaction:")
print(hghest_amount_trans(transactions))


def returning_ids(trans_lst):
    trans_ids=[]
    user_ids=[]
    for transaction in trans_lst:
        t = transaction[0]
        u = transaction[1]
        trans_ids.append(t)
        user_ids.append(u)
    return f'List of Transaction ids: {trans_ids}\nList of User ids:{user_ids}'

print()
print(returning_ids(transactions))

```

```python
#task4
feedback1 = {
    'user1': {'rating': 5, 'comments': 'Excellent'},
    'user2': {'rating': 4, 'comments': 'Very Good'},
    'user3': {'rating': 2, 'comments': 'Not Good'},
}

feedback2 = {
    'user2': {'rating': 5, 'comments': 'Great'},
    'user4': {'rating': 3, 'comments': 'Average'},
}

def filterUser(dic):
    newDic = {}
    for key, value in dic.items():
        if value['rating'] >= 4:
            newDic[key] = value['rating']

    sortedDic = dict(sorted(dic.items(), key=lambda item: item[1]['rating'], reverse=True)[:5])
    return newDic, sortedDic


filtered_users, top_usrs = filterUser(feedback1)
print("Users with rating 4 or higher:", filtered_users)
print()
print("Top 5 users by rating:", top_usrs)

def combining(*dicts):
    combinedDic = {}
    for dic in dicts:
        for key, value in dic.items():
            if key not in combinedDic:
                combinedDic[key] = value
            else:
                combinedDic[key]['rating'] = max(combinedDic[key]['rating'], value['rating'])
                combinedDic[key]['comments'] += ', ' + value['comments']

    ratedDic = {key: value['rating'] for key, value in combinedDic.items() if value['rating'] > 3}
    return combinedDic, ratedDic


combined_feedback, rated_users = combining(feedback1, feedback2)
print()
print("Combined Feedback:", combined_feedback)
print()
print("Users with rating greater than 3:", rated_users)
```
